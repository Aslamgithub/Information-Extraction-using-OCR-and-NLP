```python
import pandas as pd
import numpy as np
import re

import warnings
warnings.filterwarnings("ignore")
```

```python
# Open text file
with open(r'C:\Users\PC WORLD\OneDrive\All\AAA\Images\ALL\all_clean.txt', "r") as filename: # path of train text file
    C_text = filename.readlines()
```

```python
C_text[:]
```

```python
from pandas import DataFrame
your_list = C_text
df = DataFrame (your_list,columns=['Advertisements'])
```

```python
df.head(2)
```

```python
#Property_Type
property='House|Apartment|Villa'
def split_it(Number):
    return re.findall(property,Number, flags=re.IGNORECASE )
df['Property_Type'] = df['Advertisements'].apply(split_it)

df['Property_Type'] = df['Property_Type'].str.get(0)

df['Property_Type'].head(2)
```

```python
#Address
def split_it(Number):
    
    return re.findall(r'(?:province|district)[\.\s]?[\s\.]?\d?\s?(\w+\s?\d?)|\spm\s(\w+\s?\w+)|forsale(\w+)', Number)
    

df['Address'] = df['Advertisements'].apply(split_it)

df['Address'] = df['Address'].str.get(0)
df['Address'].head()
```

```python
#Advertiser
def split_it(Number):
    
    return re.findall(r'negotiable\s(\w+\.?\s\w+)|(\w+\.?\s\w+)\.{3}\s|contact\sname\s(\w+\s\w+)|published\s(\D+)', Number)
    

df['Advertiser'] = df['Advertisements'].apply(split_it)

df['Advertiser'] = df['Advertiser'].str.get(0)
df['Advertiser'].head()
```

```python
#Contact
def split_it(Number):
    #return re.findall(r'0\d{7,10}|7[0-9]{8}', Number)
    return re.findall(r'\b0[71]?\d{8}|\b7\d{8}[^0+]|947\d{8}', Number) #0754012392
    #return re.findall(r'\b7\d{8}[^0+]}', Number)

df['Contact'] = df['Advertisements'].apply(split_it)

df['Contact'] = df['Contact'].str.get(0)
df['Contact'].head(2)
```

```python
#Area_of_Land
def split_it(Number):
    return re.findall(r'\d{0,}\.?\d{1,}\sperch|\d+\sp\s', Number,flags=re.IGNORECASE)

df['Area_of_Land'] = df['Advertisements'].apply(split_it)

df['Area_of_Land'] = df['Area_of_Land'].str.get(0)
df['Area_of_Land'].head(2)
```

```python
#Property_size
def split_it(Number):
    return re.findall(r'\d\,?\d+\.?\d+?\ss[iq]+[ft]?|\d+\ssq|\d+\,\d+\ssqft', Number,flags=re.IGNORECASE)
    
    

df['Property_size'] = df['Advertisements'].apply(split_it)

df['Property_size'] = df['Property_size'].str.get(0)
df['Property_size'].head(2)
```

```python
#Bedrooms
def split_it(Number):
    return re.findall(r'(\s\d{1,2}\sbedroom)|(bedroom\s\d+?)\sbathroom\s[\d+?]|(bedroom\s\d+?)\s[^b]', Number,flags=re.IGNORECASE)

df['Bedrooms'] = df['Advertisements'].apply(split_it)

df['Bedrooms'] = df['Bedrooms'].str.get(0)
df['Bedrooms'].head(2)
```

```python
#Bathrooms
def split_it(Number):
    #return re.findall(r'(\d+)(?=\s*Bathroom)', Number,flags=re.IGNORECASE)
    return re.findall(r'(?:\d\sbedroom|[^m])(\s\d\sbathroom)|(bathroom\s\d)', Number,flags=re.IGNORECASE)

df['Bathrooms'] = df['Advertisements'].apply(split_it)

df['Bathrooms'] = df['Bathrooms'].str.get(0)
df['Bathrooms'].head(2)
```

```python
#Price
def split_it(Number):
    #return re.findall(r'(\d+)(?=\s*Bathroom)', Number,flags=re.IGNORECASE)
    return re.findall(r'\sr\.\s\d+\s?\d+|rupees\s(\d+\,?\s?\d+\,?\s?\d+)|\d+\s?[0]{4,}|\d+\,\d+\,\d+', Number,flags=re.IGNORECASE)

df['Price'] = df['Advertisements'].apply(split_it)

df['Price'] = df['Price'].str.get(0)
df['Price'].head(2)
```
