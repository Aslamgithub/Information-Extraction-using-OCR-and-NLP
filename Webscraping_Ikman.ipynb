```python

baseurl =  "https://ikman.lk"


headers= {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'
}
productlinks=[]
data=[]


for x in range (1,5):
    r = requests.get(f'https://ikman.lk/en/ads/sri-lanka/houses-for-sale?sort=date&order=desc&buy_now=0&urgent=0&page={x}')
    soup= BeautifulSoup(r.content,'lxml')
 
    
        
    
    
    productlist=soup.find_all("li",class_="normal--2QYVk gtm-normal-ad")
        

           
    for item in productlist:
        for link in item.find_all('a',href=True):
            productlinks.append(baseurl + link['href'])

#print(len(productlinks))
#print(productlinks)

for link in productlinks:
    
    r=requests.get(link,headers=headers)

    soup= BeautifulSoup(r.content,'lxml')
    
    
    try:

        Address=[]
        Area_of_Land=[]
        Property_size=[]

        Bedrooms= soup.findAll('div',class_='word-break--2nyVq value--1lKHt',id= False)[1].text
        
        Bathrooms= soup.findAll('div',class_='word-break--2nyVq value--1lKHt',id= False)[2].text
        
        Advertiser= soup.find('div',class_='ellipsis--3FPDI contact-name--m97Sb').text.strip()
        
        Price= soup.find('div',class_='amount--3NTpl').text.strip()
        
        
        for item in soup.findAll(text=re.compile('Address:')):
            Address.append(item.find_next('div').text) 

        for item in soup.findAll(text=re.compile('Land size:')):
            Area_of_Land.append(item.find_next('div').text)
            
            
        for item in soup.findAll(text=re.compile('House size:')):
            Property_size.append(item.find_next('div').text)
                    
    except:
        Address=''
        Area_of_Land=''
        Bedrooms=''
        Bathrooms=''
        Property_size=''
        Advertiser=''
        Price= ''    
             
    house={
        'Address':Address,
        'Area_of_Land':Area_of_Land,
        'Bedrooms':Bedrooms,
        'Bathrooms':Bathrooms,
        'Property_size':Property_size,
        'Advertiser':Advertiser,
        'Price':Price   
    }
    
    data.append(house)
        
df = pd.DataFrame(data)

df.insert(0, 'Property_type', 'House')

df['Address'] = df['Address'].str.get(0)
df['Area_of_Land'] = df['Area_of_Land'].str.get(0)
df['Property_size'] = df['Property_size'].str.get(0)



df = df.dropna(how='any')
df = df.reset_index(drop=True)
df
```

```python
df.Address = df.Address.apply(lambda x: np.nan if 'All' in x else x)
df.Property_size = df.Property_size.apply(lambda x: np.nan if 'All' in x else x)
df.Area_of_Land = df.Area_of_Land.apply(lambda x: np.nan if 'All' in x else x)
df
```
